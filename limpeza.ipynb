```python
import numpy as np
import pandas as pd
```

data = np.loadtxt('./dadosTreinamentoClasses.csv.', skiprows=1, delimiter=',')

```python
df = pd.read_csv('./dadosTreinamentoClasses.csv');
```

```python
df[:4]
```

```python
genes = df[df.columns[:-1]]
```

```python
genesThreshold = genes.clip(upper=16000, lower=20)
```

```python
genesFoldDiff = genesThreshold.loc[:,(genesThreshold.max(axis=0) - genesThreshold.min(axis=0)) > 2]
```

```python
genesFoldDiff[:3]
```

```python
dadosTrainClassesLimpos = genesFoldDiff
```

```python
dadosTrainClassesLimpos['Class'] = df['Class']
```

```python
dadosTrainClassesLimpos[:5]
```

```python
dadosTrainClassesLimpos.to_csv('dadosTreinamento_Limpos.csv', index=False)
```

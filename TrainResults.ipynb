##Resultados do treinamento

```python
import pandas as pd
import numpy as np
```

Os treinamentos foram realizados 10 vezes cada, sendo que cada vez foi realizada a validação cruzada 5-folds

```python
 teste = pd.read_csv('./trainResults_1.csv');
```

```python
mediaTeste1 = teste.groupby(['Key_Dataset','Key_Scheme','Key_Scheme_options']).aggregate(np.mean)
```

```python
results = mediaTeste1[['Percent_correct']]
```

Apresentamos os valores médios de acurácia:

```python
results
```

Os 5 melhores

```python
results.sort('Percent_correct', ascending=False)[:5]
```

O melhor fica sendo 1bk com k igual 4 para os atributos correspondentes ao arquivo train_top4 ou train_top25
